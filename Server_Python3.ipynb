```python
import threading
from queue import Queue
import time
import socket

PORT = 10067
REGISTROS = dict()
```

```python
def threader( tiempoEjecucion ):
    while True:
        # gets an worker from the queue
        tiempoEjecucion = q.get()
        # Run the example job with the avail worker in queue (thread)
        ejecutarPaquete(tiempoEjecucion)
        # completed with the job
        q.task_done()
    print("\n Threader")
    
def ejecutarPaquete(tiempoEjecucion ):
    time.sleep(tiempoEjecucion)
    print("\n Ejecutar Paquete")
    
```

```python
# Create the queue and threader 
q = Queue()
# how many threads are we going to allow for
for x in range(30):
     t = threading.Thread(target=threader)

     # classifying as a daemon, so they will die when the main dies
     t.daemon = True

     # begins, must come after daemon definition
     t.start()
# wait until the thread terminates.
```

```python
import socket
import sys
import time
from threading import Thread 


class ClientThread(): 
 
    def __init__(self,conn, ip,port): 
        Thread.__init__(self) 
        self.ip = ip 
        self.port = port 
        self.conn=conn
        #print "[+] New server socket thread started for " + ip + ":" + str(port) 
 
    def run(self): 
        a=1
        while a==1:
            data =  self.conn.recv(2048).decode('utf-8')
            data = data.split(" ")
            user = data[0]
            userPack =  data[2]
            userPackLen = int(data[4])
            
            print( "\n \t *** {} {} {} ***".format( user, userPack, userPackLen))
            
            #Se necesita llevar un registro de los paquetes procesados para cada cliente
            try: 
                print("Returning Client: ".format( REGISTROS[user] ))
                REGISTROS[user][userPack] = userPackLen
            except Exception as e:
                print("Nuevo cliente: {}".format(user))
                user = data[0]
                REGISTROS[user] = dict()
                REGISTROS[user][userPack] = userPackLen
                
            
            
            #print ("\nServer received:   "+ data)
            MESSAGEINICIO = "\n\tSERVIDOR: INICIANDO (self.port  "+str(self.port)+"): "+ user                  
            MENSAJECORRIENDO = "\n\t------SERVIDOR: CORRIENDO ("+str(self.port)+")" 
            MESSAGEFINAL = "\n\t\tSERVIDOR: TERMINANDO ("+str(self.port)+"): "+ user
            if MESSAGEFINAL == 'exit':
                break
            try:
                print(MESSAGEINICIO)
                self.conn.send(str.encode(MESSAGEINICIO) ) # echo
                print(MENSAJECORRIENDO)
                self.conn.send(str.encode(MENSAJECORRIENDO ) ) # echo                
                
                #-------------------------THREADING---------------------------
                #el input debera camboarse por el tiempo de ejecucion de cada paquete. 
                q.put(userPackLen)

                
                print(MESSAGEFINAL)
                self.conn.send(  str.encode(MESSAGEFINAL) )  # echo 
                self.conn.close()
                a=0
            except Exception as e:
                a=0
                print("ERROR: {}".format(e))
                


```

```python
# Create a TCP/IP socket
sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
# Bind the socket to the port
server_address = ('localhost', PORT)
#print >>sys.stderr, 'starting up on %s port %s' % server_address
sock.bind(server_address)

# Listen for incoming connections
inicio=time.time()

while True:
    sock.listen(100)
    (conn, (ip,port)) = sock.accept()
    newthread = ClientThread(conn, ip,port) 
    newthread.run() 
    tiempo = time.time()
    if tiempo-inicio > 100:
        break
```
