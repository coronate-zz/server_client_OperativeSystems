```python
import socket
import sys
import time
from threading import Thread 


class ClientThread(): 
 
    def __init__(self,conn, ip,port): 
        Thread.__init__(self) 
        self.ip = ip 
        self.port = port 
        self.conn=conn
        #print "[+] New server socket thread started for " + ip + ":" + str(port) 
 
    def run(self): 
        a=1
        while a==1:
            data =  self.conn.recv(2048) 
            print ("\nServer received:   "+ data.decode('utf-8'))
            MESSAGEINICIO = "\n\tSERVIDOR: INICIANDO (self.port  "+str(self.port)+"): "+ data.decode('utf-8')
            MESSAGEFINAL = "\n\t\tSERVIDOR: TERMINANDO ("+str(self.port)+"): "+ data.decode('utf-8')
            if MESSAGEFINAL == 'exit':
                break
            try:
                print("\nTrying to send " + MESSAGEINICIO)
                self.conn.send(str.encode(MESSAGEINICIO) ) # echo
                self.conn.send(str.encode( "\n\t------SERVIDOR: CORRIENDO ("+str(self.port)+")") ) # echo
                time.sleep(1)
                print("\nTrying to send " + MESSAGEFINAL)
                self.conn.send(  str.encode(MESSAGEFINAL) )  # echo 
                self.conn.close()
                a=0
            except:
                a=0
                print("ERROR ERROR ERROR")


```

```python
# Create a TCP/IP socket
sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
# Bind the socket to the port
server_address = ('localhost', PORT)
#print >>sys.stderr, 'starting up on %s port %s' % server_address
sock.bind(server_address)

# Listen for incoming connections
inicio=time.time()

while True:
    sock.listen(100)
    (conn, (ip,port)) = sock.accept()
    newthread = ClientThread(conn, ip,port) 
    newthread.run() 
    tiempo = time.time()
    if tiempo-inicio > 30:
        break
```
