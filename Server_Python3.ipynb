```python
PORT = 10055
REGISTROS = dict()
```

```python
import socket
import sys
import time
from threading import Thread 


class ClientThread(): 
 
    def __init__(self,conn, ip,port): 
        Thread.__init__(self) 
        self.ip = ip 
        self.port = port 
        self.conn=conn
        #print "[+] New server socket thread started for " + ip + ":" + str(port) 
 
    def run(self): 
        a=1
        while a==1:
            data =  self.conn.recv(2048).decode('utf-8')
            
            #Se necesita llevar un registro de los paquetes procesados para cada cliente
            try: 
                print("Returning Client: ".format( REGISTROS[data] ))
            except Exception as e:
                print("Nuevo cliente registrado: {}".format(data))
                REGISTROS[data] = dict()
                
            
            
            #print ("\nServer received:   "+ data)
            MESSAGEINICIO = "\n\tSERVIDOR: INICIANDO (self.port  "+str(self.port)+"): "+ data
            MESSAGEFINAL = "\n\t\tSERVIDOR: TERMINANDO ("+str(self.port)+"): "+ data
            if MESSAGEFINAL == 'exit':
                break
            try:
                print("\nTrying to send " + MESSAGEINICIO)
                self.conn.send(str.encode(MESSAGEINICIO) ) # echo
                self.conn.send(str.encode( "\n\t------SERVIDOR: CORRIENDO ("+str(self.port)+")") ) # echo                
                time.sleep(1)
                print("\nTrying to send " + MESSAGEFINAL)
                self.conn.send(  str.encode(MESSAGEFINAL) )  # echo 
                self.conn.close()
                a=0
            except Exception as e:
                a=0
                print("ERROR: {}".format(e))
                


```
